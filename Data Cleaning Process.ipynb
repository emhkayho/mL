### Looking at null values

```python
#filling in null values:
from sklearn.impute import SimpleImputer
# filling in null values for categorical  data
imp_mean_cat = SimpleImputer(strategy='most_frequent')
objects[['parent_id','name','normalized_name',
         'category_code','founded_at','closed_at',
         'domain','homepage_url','twitter_username',
         'logo_url','short_description','description',
         'overview','tag_list','country_code','state_code',
         'city','first_investment_at','last_investment_at','first_funding_at','last_funding_at','first_milestone_at',
         'last_milestone_at','created_by','created_at']]= imp_mean_cat.fit_transform(objects[['parent_id','name','normalized_name','category_code',
                                                                                              'founded_at','closed_at','domain','homepage_url',
                                                                                              'twitter_username','logo_url','short_description',
                                                                                              'description','overview','tag_list','country_code',
                                                                                              'state_code','city','first_investment_at','last_investment_at','first_funding_at','last_funding_at',
                                                                                              'first_milestone_at','last_milestone_at','created_by','created_at']])
```

```python
objects.isnull().sum()
```

```python
objects_df = objects.copy()
```

```python
objects_df['first_investment_at'] = pd.to_datetime(objects_df['first_investment_at'])
objects_df['last_investment_at'] = pd.to_datetime(objects_df['last_investment_at'])
objects_df['first_funding_at'] = pd.to_datetime(objects_df['first_funding_at'])
objects_df['last_funding_at'] = pd.to_datetime(objects_df['last_funding_at'])
objects_df['first_milestone_at'] = pd.to_datetime(objects_df['first_milestone_at'])
objects_df['last_milestone_at'] = pd.to_datetime(objects_df['last_milestone_at'])
objects_df['created_at'] = pd.to_datetime(objects_df['created_at'])
objects_df['updated_at'] = pd.to_datetime(objects_df['updated_at'])
```
